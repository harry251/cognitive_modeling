```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Prior parameters
mu0 = 5.0  # Prior mean
sigma0 = 2.0  # Prior standard deviation
sigma = 1.5  # Known standard deviation of likelihood

# Simulated data (e.g., heights of a fictional species of aliens)
n = 10  # Number of observations
data = np.random.normal(loc=6.0, scale=sigma, size=n)  # Generating random data

data_mean = np.mean(data)

# Compute posterior parameters
posterior_variance = 1 / (1/sigma0**2 + n/sigma**2)
posterior_mean = posterior_variance * (mu0/sigma0**2 + n*data_mean/sigma**2)
posterior_std = np.sqrt(posterior_variance)

# Plot prior and posterior distributions
x = np.linspace(0, 10, 1000)
prior_pdf = stats.norm.pdf(x, mu0, sigma0)
posterior_pdf = stats.norm.pdf(x, posterior_mean, posterior_std)

plt.figure(figsize=(8, 5))
plt.plot(x, prior_pdf, label='Prior', linestyle='dashed')
plt.plot(x, posterior_pdf, label='Posterior', linestyle='solid')
plt.hist(data, bins=10, density=True, alpha=0.3, color='gray', label='Data Histogram')
plt.xlabel('µ (Mean)')
plt.ylabel('Density')
plt.title('Normal-Normal Model: Prior vs Posterior')
plt.legend()
plt.grid(True)
plt.show()

print(f"Prior Mean: {mu0}, Prior Std: {sigma0}")
print(f"Posterior Mean: {posterior_mean:.2f}, Posterior Std: {posterior_std:.2f}")
```
